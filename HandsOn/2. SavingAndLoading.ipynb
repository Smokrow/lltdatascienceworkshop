## Saving and Loading Experiments

An experiment framework is only useful if its records can be stored and opened again later.

Extend your framework so that an experiment can be written to disk and reconstructed from the saved data. After loading, the resulting object should contain the same relevant information as the original experiment.

### Core Challenge: JSON

Save your experiment as a JSON file and load it again.

JSON is a useful starting point because it is human-readable, widely supported, and easy to inspect with a text editor.

Your solution should:

* convert the dataclass into a JSON-compatible representation;
* write the representation to a file;
* read the file again;
* reconstruct the appropriate dataclass instances;
* verify that important values survived the round trip.

Remember that JSON only supports a limited set of data types:

* objects;
* arrays;
* strings;
* numbers;
* booleans;
* `null`.

Values such as `datetime` objects, enums, paths, sets, and custom Python objects may require an explicit conversion.

For example, a date and time could be stored as an ISO-formatted string and converted back when the file is loaded.

A successful round trip should follow this pattern:

```text
Python object → JSON-compatible data → JSON file
JSON file → JSON-compatible data → Python object
```

### Advanced Challenge: NumPy Data

NumPy arrays cannot be written directly to JSON using Python’s standard JSON encoder.

Decide how your framework should serialize numerical arrays.

One simple approach is to convert an array into a nested Python list before saving:

```python
array.tolist()
```

When loading the experiment, the list can be converted back into an array:

```python
np.asarray(loaded_values)
```

However, storing only the values may not preserve all relevant information. Consider also recording:

* the array shape;
* the data type, such as `float32` or `uint8`;
* the physical unit;
* the meaning of each dimension;
* axis labels or coordinates;
* the name of the measurement.

For larger arrays, JSON may become inefficient. As an additional challenge, consider storing metadata in JSON while saving numerical data separately.

Your experiment file could then contain a reference to the external data file.

For example:

```text
experiment.json
measurements.csv
```

Think about how these files remain connected if they are copied, renamed, or moved.

### Professional Challenge: Images

Image data introduces the same serialization questions as NumPy arrays, often at a much larger scale.

Decide whether your experiment format should store:

* the image directly;
* an encoded version of the image;
* the image as a NumPy array;
* a path to an external image file;
* a structured image record containing both a path and metadata.

Storing an entire image inside JSON is usually possible only after encoding it as text, for example using Base64. This can make JSON files very large and difficult to inspect.

A more practical approach is often to save images as separate files and reference them from the experiment metadata.

For example:

```text
experiment.json
images/
    full_cake.png
    cross_section.png
    crumb_detail.tiff
```

An image record might contain:

* a relative file path;
* the image format;
* the width and height;
* the number of channels;
* the pixel data type;
* the acquisition time;
* the camera or instrument;
* a description of the image;
* its role in the experiment.

Use relative paths where possible so that the complete experiment directory can be moved to another computer.

### Questions to Consider

While implementing saving and loading, discuss the following:

* Which values can be stored directly in JSON?
* Which values need to be converted?
* Where should serialization logic live?
* How can the file format be validated?
* What should happen if a required field is missing?
* What should happen if an older file uses a different structure?
* Should the file contain a format-version number?
* How can you verify that saving and loading did not change the data?
* Should arrays and images be embedded or stored in separate files?
* How can paths remain valid when an experiment is moved?

# Introduction to JSON

## What is JSON?

**JSON (JavaScript Object Notation)** is a lightweight, text-based data format used to store and exchange data between servers and web applications. It's easy for humans to read and write, and easy for machines to parse and generate.

---

## Why Use JSON?

- **Lightweight** – Minimal overhead compared to other formats
- **Language-independent** – Works with most programming languages
- **Easy to read** – Simple, clean structure
- **Widely supported** – Used in APIs, databases, and config files

---

## JSON Structure

JSON is built on two main structures:

1. **Objects** – Key-value pairs wrapped in curly braces `{}`
2. **Arrays** – Ordered lists of values wrapped in square brackets `[]`

### Example
General structure

```json
{
  "name": "Alice",
  "age": 28,
  "isStudent": false,
  "courses": ["Math", "Science", "History"],
  "address": {
    "city": "New York",
    "zip": "10001"
  }
}
```
Python Library

```python
import json
data = {"name": "alice", "measurement":1}
json.dumps(data)
```
---

## Data Types

| Type | Example |
|------|---------|
| String | `"Hello"` |
| Number | `42`, `3.14` |
| Boolean | `true`, `false` |
| Null | `null` |
| Array | `["a", "b"]` |
| Object | `{"key": "value"}` |

---

## Common Uses

- **API responses** – Most modern web APIs return data in JSON
- **Configuration files** – Many tools (e.g., VS Code, npm) use `config.json`
- **Data storage** – Storing structured data in files or databases

---

## Summary

JSON is the backbone of modern web development. Its simplicity and flexibility make it the go-to format for transmitting data between clients and servers. Once you understand its structure, you'll see JSON everywhere!
