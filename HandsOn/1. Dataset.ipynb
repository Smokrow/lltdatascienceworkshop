# Building an Experiment Framework with Python Dataclasses

Welcome to the **Institute for Cake Observation**.

In this workshop, you will design a small but structured experiment framework for documenting cake observations. The goal is not only to store measurement results, but also to think carefully about the context in which those results were produced.

A useful experiment record should answer questions such as:

* What was measured?
* Which sample was observed?
* Who performed the experiment?
* When and under which conditions was it conducted?
* Which settings, instruments, or methods were used?
* What data was produced?
* How can the experiment be reproduced later?

You will use Python `dataclasses` to create clear and reusable data structures for this information.

## Your Task

Design one or more dataclasses that represent a cake observation experiment.

Your framework should distinguish between:

### Experiment metadata

Metadata describes the experiment and its context.

Possible examples include:

* experiment ID;
* sample or cake name;
* observer;
* date and time;


### Experiment data

Experiment data contains the actual results of the observation.

Possible examples include:

* cake height;
* mass;
* moisture;
* number of layers;


There is no single correct solution. Choose fields and data types that make sense for your experiment, and consider which information is required, optional, or given a default value.

Extend your framework to support measurements stored as NumPy arrays.

Examples might include:

Consider how array shape, units, and measurement meaning should be documented.

### Professional Challenge: Image Data

Extend the experiment framework so that an experiment can also include one or more images.

Possible examples include:

Think about whether your dataclass should store the image itself, a NumPy array, a file path, or a separate image record containing additional metadata.


## Questions to Consider

While designing your framework, discuss the following:

* Which fields belong to metadata, and which belong to experiment data?
* Which fields should be mandatory?
* Which fields should have default values?

* How should units be represented?
* How should missing values be handled?

```python
# Example for a dataclass

from dataclasses import dataclass

@dataclass
class InventoryItem:
    """Class for keeping track of an item in inventory."""
    name: str
    unit_price: float
    quantity_on_hand: int
    broken: bool
```

```python
# Example creation of a dataclass
item1 = InventoryItem(name="Food",unit_price=3,quantity_on_hand=4, broken=False)
```
