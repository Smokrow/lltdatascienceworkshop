```python
import numpy as np
import pyvista as pv
from pyvista import examples

# Download the dataset
vol = examples.download_knee_full()

# Camera position
cpos = [
    (-381.74, -46.02, 216.54),
    (74.8305, 89.2905, 100.0),
    (0.23, 0.072, 0.97),
]

# Create the plotter
plotter = pv.Plotter()

# Initial isovalue
iso = (vol.get_data_range()[0] + vol.get_data_range()[1]) / 2

# Add the initial contour
contour = vol.contour([iso])
actor = plotter.add_mesh(contour, color="white")

def update_iso(value):
    """Update the contour for the selected isovalue."""
    global actor
    new_contour = vol.contour([value])

    # Remove the old mesh and add the new one
    plotter.remove_actor(actor)
    actor = plotter.add_mesh(new_contour, color="white")

    plotter.render()

# Add the slider
plotter.add_slider_widget(
    callback=update_iso,
    rng=vol.get_data_range(),
    value=iso,
    title="Iso Value",
)

plotter.camera_position = cpos
plotter.show(jupyter_backend="server")
```
