```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [12, 18, 14, 25, 22]

plt.figure(figsize=(9, 5))

plt.plot(
    days,
    sales,
    marker="o",
    linewidth=3,
    markersize=8,
    label="Sales",
)

plt.fill_between(days, sales, alpha=0.2)

plt.title("Weekly Sales", fontsize=16, fontweight="bold")
plt.xlabel("Day")
plt.ylabel("Units Sold")
plt.grid(True, alpha=0.3)
plt.legend()

for day, value in zip(days, sales):
    plt.annotate(
        str(value),
        (day, value),
        textcoords="offset points",
        xytext=(0, 8),
        ha="center",
    )

plt.tight_layout()
plt.show()
```
