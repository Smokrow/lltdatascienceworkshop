```python
import plotly.graph_objects as go

x = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [12, 18, 14, 25, 22]

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=x,
        y=sales,
        mode="lines+markers",
        name="Sales",
        line=dict(width=3),
        marker=dict(size=10),
        fill="tozeroy",
        hovertemplate="<b>%{x}</b><br>Sales: %{y}<extra></extra>",
    )
)

fig.update_layout(
    title="Weekly Sales",
    xaxis_title="Day",
    yaxis_title="Units Sold",
    template="plotly_white",
    hovermode="x unified",
    width=800,
    height=450,
)

fig.show()
```
