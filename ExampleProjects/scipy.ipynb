```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d

rng = np.random.default_rng(42)
x = np.linspace(0, 4 * np.pi, 200)
signal = np.sin(x) + rng.normal(0, 0.35, x.size)

filtered = uniform_filter1d(signal, size=11)

plt.plot(x, signal, label="Noisy signal", alpha=0.6)
plt.plot(x, filtered, label="Mean-filtered signal", linewidth=2)
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.title("Mean filter using SciPy")
plt.legend()
plt.tight_layout()
plt.show()
```
