```python
import numpy as np
import matplotlib.pyplot as plt
import pyswarms as ps
import ipywidgets as widgets
from IPython.display import display

def score(x):
    return -(x - 1.5) ** 2 + 4

def cost(particles):
    return -score(particles[:, 0])  # PySwarms minimizes

optimizer = ps.single.GlobalBestPSO(
    n_particles=20,
    dimensions=1,
    options={"c1": 0.5, "c2": 0.3, "w": 0.9},
    bounds=(np.array([-3.0]), np.array([5.0])),
)

optimizer.optimize(cost, iters=50, verbose=False)
history = np.asarray(optimizer.pos_history)[:, :, 0]

best_positions, best_scores = [], []
best_x, best_y = history[0, 0], -np.inf
for positions in history:
    scores = score(positions)
    i = np.argmax(scores)
    if scores[i] > best_y:
        best_x, best_y = positions[i], scores[i]
    best_positions.append(best_x)
    best_scores.append(best_y)

grid = np.linspace(-3, 5, 400)

def show_iteration(iteration):
    positions = history[iteration]
    scores = score(positions)

    plt.figure(figsize=(9, 5))
    plt.plot(grid, score(grid), label="Objective function")
    plt.scatter(positions, scores, s=70, label="Particles")
    plt.scatter(
        best_positions[iteration],
        best_scores[iteration],
        marker="*",
        s=250,
        label="Best found",
        zorder=3,
    )
    plt.title(
        f"Iteration {iteration + 1} | "
        f"Best x={best_positions[iteration]:.4f}, "
        f"score={best_scores[iteration]:.4f}"
    )
    plt.xlabel("x")
    plt.ylabel("Score")
    plt.xlim(-3, 5)
    plt.ylim(score(grid).min() - 1, score(grid).max() + 1)
    plt.grid(alpha=0.25)
    plt.legend()
    plt.show()

slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(history) - 1,
    step=1,
    description="Iteration:",
    continuous_update=True,
)

widgets.interact(show_iteration, iteration=slider)
```
